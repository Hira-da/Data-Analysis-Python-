```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
hr=pd.read_csv("HRDataset_v14.csv")
hr.head()
```

```python
hr.describe()
```

```python
hr.info()
```

```python
df=hr.dropna()
```

```python
#  Distribution based on gender (Sex)
plt.pie(df['Sex'].value_counts(), labels=['', ''], autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
# Customize the chart
plt.title('Gender Distribution in the Organization')
plt.legend(['Female','Male'])
# Show the chart
plt.show()
```

```python
colors = [  'lightgreen', 'lightsteelblue','lightsalmon', 'lightpink','lightgray' ]
plt.barh(df['RaceDesc'].value_counts().index, df['RaceDesc'].value_counts(),color=colors)
plt.title('Workforce Distribution by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()
```

```python
colors = ['lightsteelblue','lightsalmon', 'lightpink','lightgray']
# Plot the distribution of employees by employment status
employment_status_distribution = df['EmploymentStatus'].value_counts()
employment_status_distribution.plot(kind='bar', color=colors)
plt.title('Distribution of Employees by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Count')
# Tilt x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()

```

```python
# Question 3.1: Distribution of employees based on performance scores (PerformanceScore)
performance_distribution = df['PerformanceScore'].value_counts()
performance_distribution.plot(kind='barh', color=colors)
plt.title('Distribution of Employees by Performance Score')
plt.xlabel('Performance Score')
plt.ylabel('Count')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
sns.boxplot(x='Department', y='Salary', data=df, palette='Pastel1')
plt.title('Salary Distribution by Department')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Question 5.2: Distribution of performance scores for employees reporting to different managers
plt.figure(figsize=(12, 6))
sns.countplot(x='ManagerName', hue='PerformanceScore', data=df, palette='Pastel1')
plt.title('Distribution of Performance Scores by Manager')
plt.xlabel('Manager')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Question 6.1: Most commonly used recruitment sources
plt.figure(figsize=(12, 6))
sns.countplot(x='RecruitmentSource', data=df, palette='Pastel1', order=df['RecruitmentSource'].value_counts().index)
plt.title('Recruitment Source Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Question 7.1: Distribution of employees based on marital status
marital_status_distribution = df['MarriedID'].value_counts()
marital_status_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, labels=(["",""]),colors=['lightgray','lightsalmon'])
plt.title('Distribution of Employees by Marital Status')
plt.legend(["Single","Married"])
plt.show()
```
